The Pandas Library is a useful tool that enables us to read various datasets into a data frame; our Jupyter notebook platforms have a built-in <b>Pandas Library</b> so that all we need to do is import Pandas without installing. 

```python
import pandas as pd
```

<h2>Reading Data</h2>
<p>
Using <code>pandas.read_csv()</code> function to read the csv file. 
Because the data does not include headers, I have added an argument <code>headers = None</code>  inside the  <code>read_csv()</code> method, so that pandas will not automatically set the first row as a header.<br>
</p>
<p> This dataset is hosted on IBM Cloud object.  </p>

```python
# Reading the online file by the URL, and assigning it to variable "df"
URL_file = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv"
df = pd.read_csv(URL_file, header=None)
```

```python
#viewing the top rows of the dataset
df.head(10)
```

```python
#viewing the bottom rows of the dataset 
df.tail(10)
```

### Adding Headers to the Dataset

```python
#creating a list containing headers
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("Headers :\n", headers)
```

```python
#adding headers to our dataset
df.columns = headers
```

```python
#viewing the changes made
df.head()
```

### Saving the Dataset

```python
df.to_csv("/Users/saumyasingh/Documents/DataScience/Automobile DataSet Analysis/automobile.csv", index = False)
```
